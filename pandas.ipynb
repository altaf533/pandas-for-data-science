```python
# Use () for methods/functions.
# Use no brackets for attributes.
# Use [] for selecting data, like df["name"]
```

```python
titanic.info()
```

```python
titanic.columns
```

```python
age = titanic["age"]
age.head()
```

```python
age_sex = titanic[["age", "sex"]]
age_sex.head()
```

```python
above_35 = titanic[titanic["age"] > 35]     # passengers older than 35 years
above_35.head()
```

```python
above_35.shape
```

```python
titanic.columns
```

```python
class_23 = titanic[titanic["pclass"].isin([2, 3])]       # passengers in class 2 or 3
class_23
```

```python
class_23 = titanic[(titanic["pclass"] == 2) | (titanic["pclass"] == 3)]
class_23.head()
```

```python
class_23 = titanic[(titanic["pclass"] == 2) | (titanic["pclass"] == 3)]
class_23
```

```python
age_no_na = titanic[titanic["age"].notna()]
age_no_na
```

```python
# iloc uses integer-based indexing, so it selects rows 9 to 24 and columns 2 to 4 (since the end index is exclusive).
titanic.iloc[9:25, 2:5] 
```
